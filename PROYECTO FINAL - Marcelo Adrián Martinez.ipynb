# **MODELO DE MACHINE LEARNING**

### **√ÅRBOLES DE REGRESI√ìN**

1. Definici√≥n de variable Objetivo y variables independientes:

```python
# ==============================================================================
X = Clinica1.drop(['duracionEstancia'], axis=1)
y = Clinica1['duracionEstancia']
```

2. Verificaci√≥n de dimensiones:

```python
# ==============================================================================
print(X.shape)  # Dimensiones de X
print(y.shape) # Dimensiones de y
```

3. Divisi√≥n de datos de entrenamiento y de test:

```python
# ==============================================================================
X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y,
                                        test_size=0.30,
                                        random_state = 0
                                    )

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

4. Creaci√≥n del modelo:

```python
# ==============================================================================
modelo = DecisionTreeRegressor(max_depth=3)
```

5. Entrenamiento del modelo:

```python
# ==============================================================================
modelo = modelo.fit(X_train, y_train)
```

6. Visualizaci√≥n del entrenamiento del modelo:

```python
fig = plt.figure(figsize=(20, 18))

tree.plot_tree(modelo, feature_names= list(X.columns.values),
                        class_names=list(y.values), filled=True)

plt.show()
```

7. Predicciones:

```python
# ==============================================================================
y_train_pred = modelo.predict(X_train)
y_train_pred
```

```python
# ==============================================================================
y_test_pred = modelo.predict(X_test)
y_test_pred
```

8. Comparativa con las etiquetas reales:

```python

print('Error en datos de train:', mean_squared_error(y_train_pred,y_train))
print('Error en datos de test:', mean_squared_error(y_test_pred,y_test))
print('Score en datos de train:', modelo.score(X_train, y_train)*100)
print('Score en datos de test:', modelo.score(X_test, y_test)*100)

ad_test_mse = mean_squared_error(y_train, y_train_pred)
ad_train_mse = mean_squared_error(y_test, y_test_pred)
```

9. M√©tricas

```python
#===============================================================================

mse_ad = mean_squared_error(y_test, y_test_pred)
print("MSE:", mse_ad)
#--------------------------------------------------
rmse_ad = np.sqrt(mse_ad)
print("RMSE:", rmse_ad)
#--------------------------------------------------
mae_ad = MAE(y_test, y_test_pred)
print("MAE:", mae_ad)
#--------------------------------------------------
r2_ad = r2_score(y_test, y_test_pred)
print("R^2:", r2_ad)
```

### **RANDOM FOREST**

1. Divisi√≥n de los datos en entrenamiento y test:

```python
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
                                        Clinica1.drop(columns = "duracionEstancia"),
                                        Clinica1['duracionEstancia'],
                                        random_state = 123
                                    )
```

2. Creaci√≥n del modelo:

```python
# ==============================================================================
modelo = RandomForestRegressor(
            n_estimators = 200,
            criterion    = "squared_error",
            max_depth    = 6,
            max_features = 'sqrt',
            oob_score    = True,
            n_jobs       = -1,
            random_state = 123
         )
```

3. Entrenamiento del modelo:

```python
# ==============================================================================
modelo.fit(X_train_rf, y_train_rf)
```

4. Predicciones:

```python
# ==============================================================================
rf_predic_y_train = modelo.predict(X_train_rf)
rf_predic_y_test  = modelo.predict(X_test_rf)
```

5. Error de test del modelo inicial:

```python
# ==============================================================================

print('Error en datos de train:', mean_squared_error(rf_predic_y_train, y_train_rf))
print('Error en datos de test:', mean_squared_error(rf_predic_y_test,y_test_rf))
print('Score en datos de train:', modelo.score(X_train_rf, y_train_rf)*100)
print('Score en datos de test:', modelo.score(X_test_rf, y_test_rf)*100)

rfr_test_mse = mean_squared_error(y_train_rf, rf_predic_y_train)
rfr_train_mse = mean_squared_error(y_test_rf, rf_predic_y_test)
```

6. M√©tricas

```python
#===============================================================================

mse_rf = mean_squared_error(y_test_rf, rf_predic_y_test)
print("MSE:", mse_rf)
#--------------------------------------------------
rmse_rf = np.sqrt(mse_rf)
print("RMSE:", rmse_rf)
#--------------------------------------------------
mae_rf = MAE(y_test_rf, rf_predic_y_test)
print("MAE:", mae_rf)
#--------------------------------------------------
r2_rf = r2_score(y_test_rf, rf_predic_y_test)
print("R^2:", r2_rf)
```
